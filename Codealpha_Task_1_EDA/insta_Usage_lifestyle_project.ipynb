### Feature Engineering

```python
bins = [-1, 18, 30, 45, 60, df['age'].max()]
labels = ['<18', '18-30', '31-45', '46-60', '60+']

df['age_group'] = pd.cut(
    df['age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
### Here filling median values on the basis of their category
df["sleep_hours_per_night"] = (df.groupby("age_group",observed = True)["sleep_hours_per_night"].transform(lambda x: x.fillna(x.median())))
df["perceived_stress_score"] = (df.groupby("age_group",observed = True)["perceived_stress_score"].transform(lambda y : y.fillna(y.median())))
```

## ***Exploratory Data Analysis**

### ***1Q....How does user engagement vary across different employeement_status?***

```python
# Calculate the average user engagement score for each employment status
emp_status_grp = (
    df.groupby("employeement_status")["user_engagement_score"]
      .mean()
      .reset_index()
)

# Create a new figure with defined size and resolution for better clarity
plt.figure(figsize=(8, 5), dpi=120)

# Sort employment status categories based on average engagement (descending order)
# This ensures the bar plot shows highest engagement first
order = emp_status_grp.sort_values(
    "user_engagement_score",
    ascending=False
)["employeement_status"]

# Create a bar plot to visualize average engagement by employment status
ax = sns.barplot(
    data=emp_status_grp,
    x="employeement_status",
    y="user_engagement_score",
    palette="pastel",     # Soft color palette for clean, professional look
    alpha=0.9,            # Slight transparency to reduce visual harshness
    order=order           # Apply sorted order for better interpretability
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

# Add a descriptive title to the plot
plt.title(
    "Avg User Engagement by Employeement Status",
    fontsize=14,
    fontweight="normal",
    color="purple"
)

# Label the x-axis clearly
plt.xlabel(
    "Employeement Status",
    fontsize=12,
    fontweight="normal",
    color="black"
)

# Label the y-axis clearly
plt.ylabel(
    "Avg Engagement Score",
    fontsize=12,
    fontweight="normal",
    color="black"
)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the final plot
plt.show()
```

```python
Key Insights

1...Retired users have the highest average engagement, likely because they have more free time.

2...Full-time employees, freelancers, and students show similar and stable engagement levels.

3...Unemployed and part-time users have slightly lower engagement compared to other groups.

Business Recommendations

1...Create more informative and long-form content for retired users to keep them engaged.

2...Post short and impactful content during breaks and evenings for working professionals and students.

3...Share helpful and skill-based content to improve engagement among unemployed and part-time users.
```

### ***5Q...How are user engagement scores related to usage behavior metrics such as reels watched, likes given, comments written, and time spent on the app?***

```python
# Select numerical columns related to user engagement and activity
heatmap_cols = [
    'user_engagement_score',
    'reels_watched_per_day',
    'stories_viewed_per_day',
    'likes_given_per_day',
    'comments_written_per_day',
    'time_on_feed_per_day',
    'time_on_reels_per_day',
    'time_on_explore_per_day'
]

# Create a correlation matrix
corr_matrix = df[heatmap_cols].corr()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the heatmap
sns.heatmap(
    corr_matrix,              
    annot=True,                
    fmt=".2f",                
    cmap="magma",          
    linewidths=1.5,           
    cbar=True                  
)

# Add title
plt.title("Correlation Heatmap of User Engagement & Activity Metrics",color = "black")

# Display the heatmap
plt.show()
```

### **Hypothesis Testing – Exploratory Data Analysis**

### ***Relationship Between Screen Time and Perceived Stress***

Research Question

Does increased screen time lead to higher perceived stress levels among users?

Hypotheses

H0 (Null Hypothesis):
There is no significant relationship between screen time and perceived stress score.

H1 (Alternative Hypothesis):
There is a significant relationship between screen time and perceived stress score.

Variables Used

Independent Variable: Screen Time
(time_on_feed_per_day + time_on_reels_per_day + time_on_explore_per_day)

Dependent Variable: perceived_stress_score

```python
# Create total screen time by summing different screen time components
df['total_screen_time'] = (
    df['time_on_feed_per_day'] +
    df['time_on_reels_per_day'] +
    df['time_on_explore_per_day']
)
```

### Visualization (Scatter Plot + Trend Line)

```python
### Giving figure size
plt.figure(figsize=(8,5))


### Creating regular plot
sns.regplot(
    data=df,
    x='total_screen_time',
    y='perceived_stress_score',
    scatter_kws={'alpha':0.4},
    line_kws={'color':'red'}
)

### Title
plt.title('Screen Time vs Perceived Stress Score')

### x-axis name
plt.xlabel('Total Screen Time per Day')

### y-axis name
plt.ylabel('Perceived Stress Score')

### showing plot
plt.show()
```

### Statistical Test (Pearson Correlation)

```python
### from scipy we are importing stats 
from scipy.stats import pearsonr


### performing correlation and p_value
corr, p_value = pearsonr(
    df['total_screen_time'],
    df['perceived_stress_score']
)

print("Correlation:", corr)
print("P-value:", p_value)
```

```python
Final Decision

Since the p-value (0.000) is significantly less than 0.05, we reject the Null Hypothesis (H0).
```

```python
Interpretation

There is a strong positive relationship between total screen time and perceived stress score.

As screen time increases, perceived stress levels increase sharply

The correlation value of 0.82 indicates a very strong association

This relationship is statistically significant, not due to random chance
```

```python
Key Insight

Excessive screen exposure is strongly associated with higher stress levels among users.

Recommendations

Encourage digital well-being features such as screen-time limits and break reminders

Promote content moderation to reduce overstimulation

Introduce stress-relief or mindfulness content for high screen-time users

Educate users on maintaining a balanced digital lifestyle
```

```python
The hypothesis test reveals a strong positive correlation (r = 0.82) between total screen time and perceived stress score. 
With a p-value less than 0.05, the null hypothesis is rejected, indicating that increased screen time is significantly 
associated with higher stress levels.
```
