### ***1Q....Which product categories generate the highest total sales?***

### here coverting final price in a million value format ###

```python
cat_sales = df.groupby("category")["final_price"].sum().reset_index()
cat_sales["final_price_in_m"] = (cat_sales["final_price"]/1000000).round(1)
cat_sales
```

```python
plt.figure(figsize=(8,5))

order = cat_sales.sort_values("final_price_in_m",ascending=False)["category"]
ax = sns.barplot(data = cat_sales,x="category",y="final_price_in_m",
                 order = order,palette="viridis",alpha=0.9
                )

for container in ax.containers:
    ax.bar_label(container,fmt="%.1fm")

plt.xticks(rotation=45)
plt.title("Category Wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

 Insights

Beauty and Toys categories generate the highest sales, showing strong and consistent customer demand.

Sales across all categories are relatively close, indicating a well-balanced product portfolio.

Home & Kitchen and Appliances show slightly lower sales, suggesting comparatively weaker performance.

 Business Recommendations

Increase inventory and marketing spend for high-performing categories like Beauty and Toys to maximize revenue.

Introduce targeted promotions or bundles for Home & Kitchen and Appliances to boost sales.

Analyze customer preferences in lower-performing categories to refine pricing, discounts, or product offerings.

### ***How does product rating vary across different brands?***

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=df,x="brand",y="rating",showmeans=True,notch=False,patch_artist=True,
           boxprops=dict(color="pink"),linewidth=2,linecolor="orange")

plt.title("Rating Distribution Across Brands")
plt.xlable("Rating")
plt.ylabel("Brands")
plt.xticks(rotation=45)
plt.show()
```

The wide spread of ratings (from 1 to 5) suggests inconsistent customer experiences within brands.

No single brand clearly outperforms others in ratings, pointing to strong competition.

### Is there a relationship between product rating and units sold? ###

```python
# Set figure size
plt.figure(figsize=(7,5))

# Create scatter plot to analyze relationship
sns.scatterplot(data=df, x="rating", y="units_sold")

# Add title and axis labels
plt.title("Relationship Between Product Rating and Units Sold")
plt.xlabel("Product Rating")
plt.ylabel("Units Sold")

# Display the plot
plt.show()
```

Insights

Units sold are widely spread across all rating levels, indicating no strong direct relationship.

Products with low ratings can still achieve high sales, suggesting other factors influence purchases.

Pricing, discounts, and brand presence likely play a larger role than ratings alone.

 Business Recommendations

Do not rely only on ratings to predict sales performance; consider pricing and promotions.

Improve product visibility and marketing for well-rated products to leverage customer trust.

Analyze additional variables such as discounts and brand loyalty to optimize sales strategy.

### Which brands offer the highest average discount? ###

```python
# Group data by brand and calculate average discount
brand_discount = df.groupby("brand")["discount_percent"].mean().reset_index()

# Sort brands by discount percentage
brand_discount = brand_discount.sort_values(by="discount_percent", ascending=False)

# Set figure size
plt.figure(figsize=(9,5))

# Create bar chart
sns.barplot(data=brand_discount, x="discount_percent", y="brand")

# Add title and labels
plt.title("Average Discount Percentage by Brand")
plt.xlabel("Average Discount (%)")
plt.ylabel("Brand")

# Display the plot
plt.show()
```

Insights

Most brands offer similar average discounts, clustered around the low-20% range.

LG and Apple appear slightly more aggressive with discounts compared to other brands.

The limited variation suggests discounting is a common competitive strategy across brands.

Business Recommendations

Differentiate beyond discounts by improving product value, features, or service quality.

Use targeted, time-bound discounts instead of flat discounting to protect profit margins.

Combine discounts with loyalty programs or bundles to increase customer retention.

### How does delivery time vary across different seller cities? ###

```python
# Set figure size
plt.figure(figsize=(9,5))

# Create box plot to show delivery day distribution
sns.boxplot(data=df, x="seller_city", y="delivery_days")

# Add title and labels
plt.title("Delivery Days Distribution Across Seller Cities")
plt.xlabel("Seller City")
plt.ylabel("Delivery Days")

# Display the plot
plt.show()
```

Key Insights

All cities show a similar median delivery time of around 6 days, indicating consistent logistics performance.

The delivery range (approximately 1 to 11 days) is wide across every city, suggesting variability in fulfillment speed.

No city stands out as significantly faster or slower, implying standardized delivery operations nationwide.

 Business Recommendations

Focus on reducing delivery variability by improving last-mile logistics and courier coordination.

Introduce city-level SLAs to ensure faster deliveries for high-demand regions.

Use predictive delivery estimates to set accurate customer expectations and improve satisfaction.

### Do higher discounts lead to lower final prices? ###

```python
# Set figure size
plt.figure(figsize=(7,5))

# Create regression plot to analyze trend
sns.regplot(data=df, x="discount_percent", y="final_price")

# Add title and axis labels
plt.title("Discount Percentage vs Final Price")
plt.xlabel("Discount Percentage")
plt.ylabel("Final Price")

# Display the plot
plt.show()
```

Insights

There is a clear negative relationship between discount percentage and final price.

Higher discounts (30–50%) lead to a sharp drop in final prices.

Lower or no discounts help maintain higher product value and margins.

 Recommendations

Use moderate discounts (10–20%) to balance sales growth and profitability.

Avoid heavy discounts unless clearing old or slow-moving inventory.

Apply targeted discounts based on customer segments to maximize revenue impact.
